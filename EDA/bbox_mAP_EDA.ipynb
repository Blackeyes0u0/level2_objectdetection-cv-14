```python
import mmcv
from mmdet.apis import inference_detector, init_detector
import matplotlib.pyplot as plt

# 모델과 설정 파일 불러오기
config_file = 'configs/_custom_/convnext/cascade_mask_rcnn_convnext_xlarge_patch4_window7_mstrain_480-800_giou_4conv1f_adamw_3x_coco_in22k_detr.py'
checkpoint_file = 'work_dirs/Cascade_RCNN+ConvNext_xlarge/epoch_36.pth'
model = init_detector(config_file, checkpoint_file, device='cuda:0')

# 이미지 파일 불러오기
img = mmcv.imread('/opt/ml/dataset/test/1280.jpg')

# 객체 검출 수행하기
result = inference_detector(model, img)

# 결과 시각화하기
colors = ['r', 'b', 'g', 'm', 'c']
plt.figure(figsize=(10,10))
plt.imshow(mmcv.bgr2rgb(img))
plt.axis('off')
for i, boxes in enumerate(result):
    color = colors[i % len(colors)]
    for box in boxes:
        if box[-1] > 0.5:
            bbox = box[:4]
            score = box[-1]
            plt.gca().add_patch(
                plt.Rectangle(
                    (bbox[0], bbox[1]),
                    bbox[2] - bbox[0],
                    bbox[3] - bbox[1],
                    fill=False,
                    edgecolor=color,
                    linewidth=2.5))
            plt.text(bbox[0], bbox[1] - 2,
                     f'{score:.2f}',
                     bbox=dict(facecolor=color, alpha=0.5),
                     fontsize=12, color='white')
plt.show()
```
